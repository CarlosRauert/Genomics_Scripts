```python
import networkx as nx
import pandas as pd
import multiprocessing as mp
```

```python
adj_matrix_directed = pd.read_csv("A1KU_adj_Matrix_19.csv")
adj_matrix_directed = adj_matrix_directed.drop(["Unnamed: 0"],axis=1)
adj_matrix_directed.index=adj_matrix_directed.columns
```

```python
adj_matrix_directed
```

```python
# Create directed Graph in NX

G = nx.from_pandas_adjacency(adj_matrix_directed, create_using=nx.DiGraph)
```

```python
Cycles= list(nx.simple_cycles(G))
print("\nElementary Cycles in the Graph:")
print(Cycles)
type(Cycles)
```

```python
# used ChatGPT to generate this Code.

# Inputs:
# how to enumerate all elementary paths and cycles through a graph in python?
# 


# Function to find all elementary paths between a source and target
def find_paths(args):
    G, source_node, target_node = args
    # Avoid self-loops
    if source_node != target_node:
        return list(nx.all_simple_paths(G, source=source_node, target=target_node))
    return []

# Function to parallelize the path computation
def parallel_find_paths(G):
    # Prepare arguments for each source-target pair
    all_node_pairs = [(G, source_node, target_node) for source_node in G.nodes for target_node in G.nodes]

    # Use multiprocessing to parallelize the path finding
    with mp.Pool(processes=mp.cpu_count()) as pool:
        result = pool.map(find_paths, all_node_pairs)

    # Filter out empty results and combine into a dictionary for further use
    paths_dict = {}
    for idx, (source_node, target_node) in enumerate((source, target) for source in G.nodes for target in G.nodes):
        if result[idx]:  # Add only if there are paths
            paths_dict[(source_node, target_node)] = result[idx]

    return paths_dict

# Find all elementary paths in parallel
paths_dict = parallel_find_paths(G)
```

```python
paths_list = []
values_list = list(paths_dict.values())

for i in range(len(values_list)):
    [paths_list.append(value) for value in values_list[i]]

paths_list[4]
```

```python
from collections import Counter

# Function to check if any element appears twice or more in a sublist
def has_duplicates(sublist):
    return any(count > 1 for count in Counter(sublist).values())

# Example list of sublists
lst = paths_list

# Filter all sublists that contain any duplicate elements
sublists_with_duplicates = [sublist for sublist in lst if has_duplicates(sublist)]

print(sublists_with_duplicates)
```
